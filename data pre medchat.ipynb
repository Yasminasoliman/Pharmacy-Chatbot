**Data Cleaning**

```python
import re
```

```python
df['disease_name'] = df['disease_name'].str.replace(r"\(\d+\)", "", regex=True)

df['disease_name'] = df['disease_name'].str.strip()
```

```python
df['drug_manufacturer'] = (
    df['drug_manufacturer']
    .str.replace("* Mkt:", "", regex=False)
    .str.strip()
)
```

```python
df['drug_manufacturer_origin'] = (
    df['drug_manufacturer_origin']
    .str.replace("* Country of Origin:", "", regex=False)
    .str.strip()
)
```

```python
df['generic_name'] = (
    df['generic_name']
    .str.replace("Generic Name","",regex=False)
    .str.strip()
)
```

```python
df[['disease_name',
    'drug_manufacturer',
    'drug_manufacturer_origin',
    'generic_name']].head(10)
```

**getting the number from final price **

```python
# استخراج الرقم من final_price

df['final_price_clean'] = (
    df['final_price']
    .str.extract(r'(\d+\.?\d*)')[0]
    .astype(float)
)

df[['final_price','final_price_clean']].head(10)
```

```python
# استخراج السعر الأصلي من price

df['price_clean'] = (
    df['price']
    .str.extract(r'(\d+\.?\d*)')[0]
)

df['price_clean'] = pd.to_numeric(
    df['price_clean'],
    errors='coerce'
)

df[['price','price_clean']].head(10)
```

```python
df[['final_price_clean','price_clean']].describe()
```

```python
df['knowledge_text'] = (

    "Disease: " + df['disease_name'].fillna('') +

    " | Medicine: " + df['med_name'].fillna('') +

    " | Generic: " + df['generic_name'].fillna('') +

    " | Manufacturer: " + df['drug_manufacturer'].fillna('') +

    " | Prescription: " + df['prescription_required'].fillna('Unknown') +

    " | Price: " + df['final_price_clean'].astype(str) +

    " | Description: " + df['drug_content'].fillna('')

)
```

```python
final_df = df[[
    'disease_name',
    'med_name',
    'generic_name',
    'final_price_clean',
    'prescription_required',
    'drug_manufacturer',
    'drug_content',
    'knowledge_text'
]]

print(final_df.shape)

final_df.head()
```

```python
final_df.isnull().sum()
```

```python
# تعويض النصوص
final_df['generic_name'] = final_df['generic_name'].fillna('Unknown')

final_df['prescription_required'] = (
    final_df['prescription_required']
    .fillna('Unknown')
)

final_df['drug_manufacturer'] = (
    final_df['drug_manufacturer']
    .fillna('Unknown')
)

final_df['drug_content'] = (
    final_df['drug_content']
    .fillna('No description available')
)

# تعويض السعر بالـ median
median_price = final_df['final_price_clean'].median()

final_df['final_price_clean'] = (
    final_df['final_price_clean']
    .fillna(median_price)
)
```

```python
final_df.isnull().sum()
```

**detecting noise **

```python
final_df.loc[:, 'drug_content'] = (
    final_df['drug_content']
    .str.replace(r'INTRODUCTION ABOUT',' ',regex=True)
    .str.replace(r'INTRODUCTION',' ',regex=True)
    .str.replace(r'\s+',' ',regex=True)
    .str.strip()
)
```

```python
final_df.loc[:, 'knowledge_text']=(

"Disease: "+final_df['disease_name']+

" | Medicine: "+final_df['med_name']+

" | Generic: "+final_df['generic_name']+

" | Manufacturer: "+final_df['drug_manufacturer']+

" | Prescription: "+final_df['prescription_required']+

" | Price: "+final_df['final_price_clean'].astype(str)+

" | Description: "+final_df['drug_content']

)
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np
```

```python
model = SentenceTransformer(
    'all-MiniLM-L6-v2'
)
```

```python
texts = final_df['knowledge_text'].tolist()

embeddings = model.encode(
    texts,
    show_progress_bar=True
)

print(embeddings.shape)

```

**Building vector DB**

```python
import faiss
```

```python
dimension = embeddings.shape[1]

index = faiss.IndexFlatL2(dimension)

index.add(
    np.array(embeddings).astype('float32')
)

print("Total vectors:", index.ntotal)
```

```python
def search_medicine(query, k=5):

    query_embedding = model.encode([query])

    distances, indices = index.search(
        np.array(query_embedding).astype('float32'),
        k
    )

    return final_df.iloc[
        indices[0]
    ][[
        'disease_name',
        'med_name',
        'generic_name',
        'final_price_clean'
    ]]
```

```python
search_medicine(
    "I have diabetes"
)
```

```python
search_medicine(
    "I have bacterial infection"
)
```
